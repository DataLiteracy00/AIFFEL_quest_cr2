```python
  #    은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
#여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
#객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요

#계좌 생성 클래스:
    #계좌 객체개수 (정보를 모으는)함수
    # 생성 계좌 계수 출력하는 get_account_n금 함수 depositum()함수 생성된 계좌 계수는 :0 개입니다
    #입()  최소 입금 가능 액수 1이상 #이자  5번 입금시1 프로이자
    #출금 함수 withdraw 출금은 계좌 잔고 이상 불가능
    #: Account 인스턴스에 저장된 정보를 출력하는 display_info()
    #입금과 출금 내역이 저장된 함수 eposit_history와 withdraw_history
import random


class Account:
    account_bank_total =0
    def __init__(self, name, balance):
        bank = "SC은행"
        self.bank = bank
        self.name = name
        self.account_num = ""
        self.balance = 0 + balance
        self.account_balance = format(balance, ",")
        Account.account_bank_total += 1
        self.deposit_history = []
        self.withdraw_history = []
        self.withdraw_count = 0

    def random_account_num(self):
        first = "".join([str(random.randint(0,9)) for _ in range(3)])  #for _ 에서 "_" 는 실행시포함이 안되는 의미
        middle = "".join([str(random.randint(0,9)) for _ in range(2)])
        last = "".join([str(random.randint(0,9)) for _ in range(6)])
        self.account_num = f"{first}-{middle}-{last}"

    def account_generation(self):
        print(f'생성된 계좌의 정보는 {self.bank}, 예금주 {self.name}, 계좌번호: {self.account_num}, 잔고: {self.account_balance}원입니다.')


    def account_total_display(self):

        print(f'현재 {self.bank} 총 계좌 수는: {Account.account_bank_total}개 입니다.')
        return Account.account_bank_total #클래스를 변수초기값을 두고 굳이 셀프 안쓰고 안에서 늘어난다 고하면 적용가능

    def withdraw(self):
        try:
            push_money = int(input("입금금액을 입력해주세요:"))
            if push_money > 1:
                self.balance = push_money + self.balance
                self.withdraw_history += ["".join((f'{push_money}입금'))]
                self.withdraw_count += 1
                if self.withdraw_count % 5 == 0:
                    interest = int(self.balance * 0.01) #int표현없이는 소수점 단위로표현해야 깔끔하게 나옴
                    self.balance += interest
                    print(f'이자가 {interest}원 발생했습니다.')
                print(f'현재 통장 잔고는 {format(self.balance,",")}원 입니다.')


            else:
                print("입금액은 1원보다 커야 합니다.")
        except ValueError:
            print("올바른 금액을 입력해주세요.")
        return self.balance,self.withdraw_history

    def depositum(self):
        try:
            pull_money = int(input("출금금액을 입력해주세요:"))
            if pull_money <= self.balance:
                self.balance= self.balance-pull_money
                self.deposit_history +=  ["".join((f'{pull_money}출금'))]
                print(f'현재 통장 잔고는 {format(self.balance,",")}원 입니다.')
            else:
                print("출금 금액은 원금보다 작아야합니다..")
        except ValueError:
            print("올바른 금액을 입력해주세요.")
        return self.balance,self.deposit_history

    def withdra_history(self):
        print(self.withdraw_history)
        print('-'*50)

    def depositum_history(self):
        print(self.deposit_history)
        print('-'*50)

    def account_display(self):#: Account 인스턴스에 저장된 정보를 출력하는 display_info()
        print(f"{self.bank}의 {self.name}님의 {self.account_num}계좌에는 {format(self.balance,',')}원이 있습니다. ")


a1 = Account("모두연",100000)
a2 = Account("이국한",1000000)
a3 = Account("노트북",1000000)
a1.random_account_num()
a1.account_generation()
a1.withdraw()
a1.account_total_display()
a1.withdra_history()
a1.account_display()

a2.random_account_num()
a2.account_generation()
a2.account_total_display()
a2.depositum()
a2.depositum_history()
a2.account_display()

a3.random_account_num()
a3.account_generation()
a3.account_total_display()
a3.depositum()
a3.depositum_history()
a3.account_display()

for _ in range(6):  #이자 발생하는지 예시
    a1.withdraw()


def Vip_account(*accounts):#만들어진 계좌안에서 비교해서 1000000만원이상표시
    print('-'*50)
    print('VIP 고객명단')
    print('-'*50)
    for i in accounts:

        if i.balance >= 1000000:
            print(f"예금주: {i.name}")
            print(f"잔액: {format(i.balance,',')}원")



Vip_account(a1,a2,a3)


```
